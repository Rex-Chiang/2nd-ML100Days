```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/Part01'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

#### 資料的 row 數以及 column 數

```python
row, col = app_train.shape
print("row : {}".format(row), "column : {}".format(col))
```

#### 列出所有欄位

```python
app_train.columns
```

```python
app_train.head(3)
```

#### 截取CODE_GENDERD為M之資料 (顯示前5項)

```python
app_train[app_train.loc[:,"CODE_GENDER"] == "M"].head()
```

#### 判斷哪些AMT_REQ_CREDIT_BUREAU_YEAR是遺失值 (顯示前5項)

```python
app_train[app_train.loc[:, "AMT_REQ_CREDIT_BUREAU_YEAR"].isnull()].head()
```

#### AMT_REQ_CREDIT_BUREAU_YEAR為遺失值者補 0 (顯示前5項)

```python
app_train.fillna({"AMT_REQ_CREDIT_BUREAU_YEAR": 'NaN', "AMT_REQ_CREDIT_BUREAU_YEAR": 0}).head()
```
